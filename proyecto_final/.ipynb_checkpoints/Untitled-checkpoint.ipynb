```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
%matplotlib inline

```

```python
for i in range(100,101,1):
    j =str(i).zfill(4)
    print (j)
    g=np.loadtxt('puntos/'+j+'.txt')
    b=g[0:200]
    c=g[200:500]
    g1=b.T
    g2=c.T
    x = np.array(g1[[0]])
    y = np.array(g1[[1]])
    z = np.array(g1[[2]]) 
    x1 = np.array(g2[[0]])
    y1 = np.array(g2[[1]])
    z1 = np.array(g2[[2]]) 
    fig = plt.figure()
  
    
    
    ax1 = fig.add_subplot(111, projection='3d')
    ax1.set_zlim(-500,500)
    ax1.set_xlim(-200,200)
    ax1.set_ylim(-100,100)
    ax1.scatter(x, y, z, c='g', marker='o')
    ax1.scatter(x1, y1, z1, c ='b', marker='o')
```
