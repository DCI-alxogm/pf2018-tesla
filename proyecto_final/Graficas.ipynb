```python
#se lee el documento que indica la cantidad de particulas que tiene cada aglomerado
T=np.loadtxt('tama√±o1.txt')
#Se guardan ambos numeros en dos variables 
n1=np.array(T[0])
n2=np.array(T[1])
#Se convierten los valores de las variables anteriores en valores enteros
n11=int(n1)
n22=int(n2)
nt=int(n1+n2)
```

```python
#Se lee el documento que contiene las posiciones 
g=np.loadtxt('0000.txt')
#Se define una matriz con los datos del primer aglomerado (de 0 a n1)
b=g[0:n11]
#Se define una segunda matriz con los datos del segundo aglomerado (de n1+1 a n2)
c=g[n11:nt]
d=b.T
e=c.T
```

```python
#Se define una matriz de 1 por n1 para cada posicion en "x", "y" y "z" del primer aglomerado
x = np.array(d[[0]])
y = np.array(d[[1]])
z = np.array(d[[2]]) 
#Se define una matriz de 1 por n2 para cada posicion en "x", "y" y "z" del segundo aglomerado
x1 = np.array(e[[0]])
y1 = np.array(e[[1]])
z1 = np.array(e[[2]]) 
```

```python
#
fig = plt.figure()
ax1 = fig.add_subplot(111, projection='3d')
ax1.scatter(x, y, z, c='g', marker='o')
ax1.scatter(x1, y1, z1, c ='r', marker='o')
```
